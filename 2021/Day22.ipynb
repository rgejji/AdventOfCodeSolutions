```python
import numpy as np

def readInput(input):
    start = ""
    rules = []
    lines = input.split("\n")
    for line in lines:
        firstSplit = line.split(" ")
        instruction = firstSplit[0]
        dimSplit = firstSplit[1].split(",")
        xvals = dimSplit[0].split("=")[1].split("..")
        yvals = dimSplit[1].split("=")[1].split("..")
        zvals = dimSplit[2].split("=")[1].split("..")
        rules.append([instruction, xvals, yvals, zvals])
    return rules

def calculateOnOff(rules, l=50):
    N=2*l+1
    A=np.zeros((N,N,N), dtype=np.uint8)
    shift = l
    for r in rules:
        val = 1 if r[0].lower() == "on" else 0
        xstart = shift + int(r[1][0])
        xend = shift + int(r[1][1])+1
        ystart = shift + int(r[2][0])
        yend = shift + int(r[2][1])+1
        zstart = shift + int(r[3][0])
        zend = shift + int(r[3][1])+1
        if xstart < 0 or xend >= N or ystart < 0 or yend >N or zstart < 0 or zend >N:
            #print("Warning: Out of Bounds Error, modifying instruction", r)
            xstart, xend = max(xstart, 0), min(xend, N)
            ystart, yend = max(ystart, 0), min(yend, N)
            zstart, zend = max(zstart, 0), min(zend, N)
        A[xstart:xend,ystart:yend, zstart:zend] = val
    return A

rules = readInput(input)
A = calculateOnOff(rules)
print(sum(A.flatten()))
```

```python
#This is a recusive option for part 2. Using arrays fail due to the size of the blocks, so we get rid of them and calculate recursively.

def readInputAllInts(input):
    start = ""
    rules = []
    lines = input.split("\n")
    for line in lines:
        firstSplit = line.split(" ")
        instruction = firstSplit[0]
        dimSplit = firstSplit[1].split(",")
        xvals = dimSplit[0].split("=")[1].split("..")
        yvals = dimSplit[1].split("=")[1].split("..")
        zvals = dimSplit[2].split("=")[1].split("..")
        rules.append([instruction, int(xvals[0]), int(xvals[1]), int(yvals[0]), int(yvals[1]), int(zvals[0]),int(zvals[1])])
    return rules

#Given a set of bounds and rules, returns how many spaces will be turned on via initializing
#Idea for rule that turns on:
#Number on after rule = Number on before rule + delta from rule
#Number on before rule + (number in rule area - number on in rule area before rule)
#For a rule that turns off:
#Number on after rule = Number on before rule - number on in rule area before rule
def calculateOnOffRecursive(rules, xstart, xend, ystart, yend, zstart, zend):
    if len(rules) == 0:
        return 0
    
    if xend < xstart or yend < ystart or zend < zstart:
        print("Error, invalid ordering of bounding cube")

    
    numberOn = calculateOnOffRecursive(rules[:-1], xstart, xend, ystart, yend, zstart, zend)
    
    #Check if we should apply the rule
    r = rules[-1]
    rxstart, rxend = r[1], r[2]
    rystart, ryend = r[3], r[4]
    rzstart, rzend = r[5], r[6]
    
    cxstart =  xstart if rxstart < xstart else rxstart
    cystart =  ystart if rystart < ystart else rystart
    czstart =  zstart if rzstart < zstart else rzstart
    
    cxend =  xend if rxend > xend else rxend
    cyend =  yend if ryend > yend else ryend
    czend =  zend if rzend > zend else rzend
    
    if cxend < cxstart or cyend < cystart or czend < czstart:
        return numberOn
    
    #Now that we now to apply the rule, apply the rule recursively
    xN = cxend-cxstart+1
    yN = cyend-cystart+1
    zN = czend-czstart+1
    
    numOnRuleArea = calculateOnOffRecursive(rules[:-1], cxstart, cxend, cystart, cyend, czstart, czend)
    if r[0].lower() == "on":
        return numberOn +(xN*yN*zN - numOnRuleArea)
    return numberOn - numOnRuleArea

rules = readInputAllInts(input)
#L=50
L=10000000
calculateOnOffRecursive(rules, -L, L, -L, L, -L, L)
```

```python
#First Attempt at pt 2, pt 1 code doesn't scale for 200kX200kX200k state space.
#Instead, for each cube, we find out how much on/off is done by the previous rules, in that space.
#Then we update the counts. This will be O(r^2), but have constrained space.
#This is the second attempt, which also store even the rules

import numpy as np

def getMinMaxs(rules):
    mins = [-50, -50, -50]
    maxs = [50, 50, 50]
    
    for r in rules:
        if mins[0] > int(r[1][0]):
            mins[0] = int(r[1][0])
        if mins[1] > int(r[2][0]):
            mins[1] = int(r[2][0])
        if mins[2] > int(r[3][0]):
            mins[2] = int(r[3][0])
        if maxs[0] < int(r[1][1]):
            maxs[0] = int(r[1][1])
        if maxs[1] < int(r[2][1]):
            maxs[1] = int(r[2][1])
        if maxs[2] < int(r[3][1]):
            maxs[2] = int(r[3][1])
    print("Mins:", mins, "Maxs:", maxs, "Num rules: ", len(rules))       

    
    
#assumes the bunds are lesser value followed by greater value
def calculateOnOffAlt(rules, xstart, xend, ystart, yend, zstart, zend):
    if xend < xstart or yend < ystart or zend < zstart:
        print("Error, invalid ordering of bounding cube")
    
    xN=xend-xstart+1
    yN=yend-ystart+1
    zN=zend-zstart+1
    xshift = -xstart
    yshift = -ystart
    zshift = -zstart
    
    A=np.zeros((xN,yN,zN), dtype=np.uint8)
    for r in rules:
        val = 1 if r[0].lower() == "on" else 0
        xstart = xshift + int(r[1][0])
        xend = xshift + int(r[1][1])+1
        ystart = yshift + int(r[2][0])
        yend = yshift + int(r[2][1])+1
        zstart = zshift + int(r[3][0])
        zend = zshift + int(r[3][1])+1
        
        #Fix out of bounds and if this results in order swapping, skip
        xstart, xend = max(xstart, 0), min(xend, xN)
        ystart, yend = max(ystart, 0), min(yend, yN)
        zstart, zend = max(zstart, 0), min(zend, zN)
        if xend < xstart or yend < ystart or zend < zstart:
            continue
        A[xstart:xend,ystart:yend, zstart:zend] = val
    return A
    
#getMinMaxs(rules)

totalOn = 0
rules = readInput(input)
for cnt in range(0, len(rules)):
    r = rules[cnt]
    xstart, xend = int(r[1][0]), int(r[1][1])
    ystart, yend = int(r[2][0]), int(r[2][1])
    zstart, zend = int(r[3][0]), int(r[3][1])
    numSpaces = (xend-xstart+1)*(yend-ystart+1)*(zend-zstart+1)
    
    
    #A, numSpaces = calculateOnOffSparse(rules[0:cnt], int(r[1][0]), int(r[1][1]), int(r[2][0]), int(r[2][1]), int(r[3][0]), int(r[3][1]))
    #A = calculateOnOffSparse(rules[0:cnt], -50, 50, -50, 50, -50, 50)
    A = calculateOnOffAlt(rules[:cnt], xstart, xend, ystart, yend, zstart, zend)
    numOnBeforeRule = sum(A.flatten())
    #print("Numspaces {} num on before {}".format(numSpaces, numOnBeforeRule))
    if r[0] == "on":
        totalOn += numSpaces - numOnBeforeRule
    elif r[0] == "off":
        totalOn -= numOnBeforeRule
    else:
        print("Unable to resolve rule ", r[0])
    #print("Total", totalOn)
print("Num on afterwards", totalOn)
```
