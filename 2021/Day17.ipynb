```python
import math
#Known equations using sums
#max height = yvel*(yvel+1)/2
#max dist = yvel*(yvel+1)/2
#yvel^2+yvel - maxheight*2 = 0

#If cnt = number of steps and we don't reach the end of x:
#xloc =xvel*(xvel+1)/2 - (xvel-cnt)*((xvel-cnt)+1)/2
#yloc =yvel*(yvel+1)/2 - (yvel-cnt)*((yvel-cnt)+1)/2
#So 
#yloc = maxheight - (yvel-cnt)*((yvel-cnt)+1)/2

#Given xloc,yloc:
#yloc = yvel*(yvel+1)/2 - (yvel-cnt)((yvel+1-cnt)/2
#yloc = 2yvel cnt - cnt^2 + cnt
#(2yloc/cnt) = 2yvel -cnt+1
#2yvel = cnt + (2yloc/cnt)-1
#Also
#2xvel = cnt + (2xloc/cnt)-1
#when we don't go past max x, otherwise xvel^2+xvel -2xloc = 0
#xvel = (-1 p/m sqrt(1+8xloc))/2



#So cnt | 2*yloc order to give an integer velocity. cnt may divide xloc up to a certain point, because x-acceleration drops to 0 and doesn't go negative
#Also since max height happens with the heighest yvel, and yvel increases with the heigher count, we want to find large cnts.
#So, for a given(xloc, yloc) we can find gcd, then its a matter of exhausting the square

def calcPos(xvel, yvel, numSteps):
    
    xloc = xvel*(xvel+1)/2 - (xvel-numSteps)*((xvel-numSteps)+1)/2 if numSteps <= xvel else xvel*(xvel+1)/2
    yloc =yvel*(yvel+1)/2 - (yvel-numSteps)*((yvel-numSteps)+1)/2
    return (xloc, yloc)

#Find common divisors >1
#want to ignore trivial trajectories where xloc=xvel
def findCommonDivisors(n):
    cd = []
    for i in range(abs(n),1, -1):
        if(n%i==0):
            cd.append(i)
    return cd


def testVel(xvel, yvel, d, xloc, yloc):
    pos = calcPos(xvel, yvel, d)
    if pos[0] != xloc or pos[1] != yloc or int(xvel) != xvel or int(yvel) != yvel:
        return False
    return True
    

#findBestVel returns a velocity that hits the target and achieves max height
def findBestVel(target):
    bestHeight = 0
    bestVel = None
    bestPos = None
    numSteps = None
    for xloc in range(target[0][0], target[0][1]+1):
        for yloc in range(target[1][0], target[1][1]+1):
            cds = findCommonDivisors(2*yloc)
            for d in cds:
                yvel = (d+(2*yloc/d) -1)/2
                #Try both xvel eventually decelerating to 0 and not.
                xvel = (d+(2*xloc/d) -1)/2
                if not testVel(xvel,yvel,d, xloc,yloc):
                    xvel = (-1+math.sqrt(1+8.0*xloc))/2
                    if not testVel(xvel,yvel,d, xloc,yloc):
                        continue
                maxHeight = yvel*(yvel+1)/2
                if maxHeight > bestHeight:
                    bestHeight = maxHeight
                    bestVel = (xvel, yvel)
                    numSteps = d
                    bestPos = (xloc,yloc)
                break
    return bestHeight, bestVel, numSteps, bestPos



#Perform Trajectory
#targetArea = [(235,259), (-118,-62)]
#targetArea = [(20,30), (-10,-5)]
#targetArea = [(20,29), (-10,-5)]

height, vel, numSteps, pos = findBestVel(targetArea)
print("Best height {} is achieved with velocity {} after {} steps and ending in pos {} ".format(height, vel, numSteps, pos))

#Check
print(calcPos(vel[0], vel[1], numSteps), pos)



#target area: x=235..259, y=-118..-62
```
